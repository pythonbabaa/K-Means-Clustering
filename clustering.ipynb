```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

## Load the data

```python
customer_data = pd.read_csv('/home/a/Desktop/bestsellers.csv')
```

```python
customer_data.head()
```

```python
customer_data = customer_data.drop(['Name', 'Author', 'Genre'], axis =1)
```

### Draw a scatter plot

```python
plt.scatter(customer_data.Price, customer_data.Reviews)
plt.xlabel('customer satisfaction')
plt.ylabel('customer loyalty')
plt.title('scatter plot of satisfaction and loyalty')
plt.show()
```

# Clustering

```python
cluster = KMeans(3)   # how many clusters / centroids to be made
a = cluster.fit_predict(customer_data).reshape(-1, 1)  # fit these centroids to the given data 
customer_data['cluster'] = a
plt.scatter(customer_data.Price, customer_data.Reviews, 
            c = customer_data.cluster, cmap = 'rainbow')
plt.xlabel('customer satisfaction')
plt.ylabel('customer loyalty')
plt.title('scatter plot along with clusters of satisfaction and loyalty')
plt.show()
```
